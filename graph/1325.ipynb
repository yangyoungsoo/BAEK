```python
from collections import deque

N, M = map(int, input().split())

graph = [[] for _ in range(N+1)]
queue = deque()

for i in range(M):
    connection = list(map(int, input().split()))
    A = connection[0]
    B = connection[1]

    graph[B].append(A)

def bfs(start_node, n, graph):
    count = 0
    queue = deque([start_node])
    visited = [False] * (n + 1)
    visited[start_node] = True
    
    while queue:
        curr = queue.popleft()
        for next_node in graph[curr]:
            if not visited[next_node]:
                visited[next_node] = True
                queue.append(next_node)
                count += 1 
    return count

result = []
max_hack = -1

for i in range(1, N + 1):
    cnt = bfs(i, N, graph)
    if cnt > max_hack:
        max_hack = cnt
        result = [i] 
    elif cnt == max_hack:
        result.append(i)

print(*(sorted(result)))
```
